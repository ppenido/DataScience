1) Em uma competição de salto em distância cada atleta tem direito a cinco saltos. O resultado do atleta será determinado pela média dos cinco valores restantes. Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe o nome, os saltos e a média dos saltos. O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo:

Atleta: Rodrigo
 
Primeiro Salto: 6.5 m

Segundo Salto: 6.1 m

Terceiro Salto: 6.2 m

Quarto Salto: 5.4 m

Quinto Salto: 5.3 m

Resultado final:

Atleta: Rodrigo

Saltos: 6.5 - 6.1 - 6.2 - 5.4 - 5.3

Média dos saltos: 5.9 m


```python
nome = " "
while len(nome)!=0:
    nome = input("Nome do atleta:")
    salto1 = input("Distancia do primeiro salto:")
    salto2 = input("Distancia do segundo salto:")
    salto3 = input("Distancia do terceiro salto:")
    salto4 = input("Distancia do quarto salto:")
    salto5 = input("Distancia do quinto salto:")

    print("Resultado final:")
    print("Atleta: ", nome)
    print("Saltos: ", salto1, salto2, salto3, salto4, salto5)
    print("Media:",(int(salto1) + int(salto2) + int(salto3) + int(salto4) + int(salto5))/5)
    
print("fim programa")
```

2) Qual é a diferença entre o **while** e **for**?

Resposta: Antes de entrar no laco de repeticao o while verifica uma condicao, enquanto no for nao. O for geralmente e usado quando se sabe o numero de iteracoes que serao precisas enquanto o while e usado para quando nao se sabe o numero de iteracoes. O while geralmente e usado quando se quer entrar em laco mediante a uma unica exclusiva condicao.

5) Faça um programa que receba a temperatura média de cada mês do ano e armazene-as em uma lista. Após isto, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual, e em que mês elas ocorreram (mostrar o mês por extenso: 1 – Janeiro, 2 – Fevereiro,...).

```python
meses = ["Janeiro", "Fevereiro","Marco","Abril","Maio","junho","Julho","Agosto","Setembro","Outubro","Novenbro","Dezembro"]
temperaturas = []
mediaTemp = 0
acimaMedia = []

for i in range(0,12):
  temperaturas.append(input(f"Digite a media de temperatura para o mes {meses[i]}:"))

for i in range(0,12):
  mediaTemp += int(temperaturas[i])

mediaTemp /= 12
print(f"Media da temperatura anual {mediaTemp}")
for i in range(0,12):
  if int(temperaturas[i]) > mediaTemp:
    acimaMedia.append(meses[i])

print("Mese acima da media anual: ",acimaMedia)
```

6) Crie um programa que receba o nome completo de 10 pessoas e armazene em um vetor. Depois, imprima as iniciais de cada pessoa.  Se o nome conter a palavra "de", ela deverá ser desconsiderada nas iniciais.

Exemplo:

Nome 1: Maria Aparecida Ferreira

Nome 2: Maria de Almeida

Resultado:

M.A.F

M.A


```python
def name(s): 
    l = s.split() 
    new = "" 
    for i in range(len(l)): 
        s = l[i] 
        new += (s[0].upper()+'.') 
    return new  

nomes = []
for i in range(1,3):
  nomes.append(input(f"digite o {i}º nome completo:"))

for i in range(0,2):
  nomes[i] = nomes[i].replace(" de ", " ")
  print(*name(nomes[i]))
```

7) Uma sequência de Fibonacci é a sequência inteira de 0, 1, 1, 2, 3, 5, 8....

Os dois primeiros termos são 0 e 1. Todos os outros termos são obtidos pela soma dos dois termos anteriores. Isso significa dizer que o n-ésimo termo é a soma do (n-1)º e (n-2)º termo. Crie um programa que receba a quantidade de elemento e imprima a senquencia de Fibonacci:

**Exemplo:**

Informe a quantidade de elemento: 4

Resultado:

0

1

1

2

```python
def fibonacci(x):
    a=1
    b=1
    if x==1:
        print('0')
    elif x==2:
        print('0','1')
    else:
        print('0')
        print(a)
        print(b)
        for i in range(x-3):
            total = a + b
            b=a
            a= total
            print(total)

x = int(input("Informe a quantidade de elementos:"))
fibonacci(x)
```

8) Escreva um programa que leia um número e retorne o seu fatorial.

5! = 5 x 4 x 3 x 2 x 1 = 120

9) Crie um programa que leia uma temperatura em graus Celsius ou Fahrenheit e converta para a outra escala.

**Equações:**

C =  (F - 32) * (5/9)

F = (C * 9/5) + 32


**Exemplo 1 :**

Informe a temperatura: 45F


Resultado:

A temperatura em Celsius é 7 graus.

**Exemplo 2 :**

Informe a temperatura: 7C


Resultado:

A temperatura em Fahrenheit é 45 graus.

```python
valor = input("Informe a tempereatura:")
if (valor[-1])=="F":
  valor = valor.replace("F", "")
  valor = int(valor)
  C=(int(valor)-32) * (5/9)
  print(f"A temperatura em Celsios é {C} graus.")
elif (valor[-1])=="C":
  valor = valor.replace("C", "")
  valor = int(valor)
  F=(int(valor) * 9/5) + 32
  print(f"A temperatura em Fahrenheit é {F} graus.")
```

10) Escreva um programa que leia uma sequencia de 10 números e retorne a quantidade de números pares e impares.

**Exemplo:**

Digite o 1 número: 2

Digite o 2 número: 4

Digite o 3 número: 7

Digite o 4 número: 12

Digite o 5 número: 33

Resultado:

Quantidade de números pares: 3

Quantidade de números ímpares: 2

```python
vet = []
par = 0
impar = 0

for i in range(0,10):
  vet.append(input(f"Digite o numero{i+1}:"))
  if int(vet[i])%2==0:
    par += 1
  else: 
    impar += 1
  

print("Quantidade de números pares:",par)
print("Quantidade de números impares:",impar)

```
